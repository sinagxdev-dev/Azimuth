```python
print("Select input format:")
print("1. Decimal Degrees (DD)")
print("2. Decimal-Minutes-Seconds (DMS)")

while True:
    choice = input("Enter choice (1 or 2): ")
    if choice in ['1', '2']:
        break
    print("Invalid choice. Please enter 1 or 2.")

if choice == '1':
    while True:
        azimuth = float(input("Enter azimuth in decimal degrees: "))
        if 0 <= azimuth <= 360:
            break
        print("Invalid azimuth. Must be between 0 and 360.")
else:
    while True:
        degrees = int(input("Enter degrees: "))
        minutes = int(input("Enter minutes: "))
        seconds = float(input("Enter seconds: "))
        
        azimuth = degrees + minutes/60 + seconds/3600
        
        if 0 <= azimuth <= 360 and 0 <= minutes < 60 and 0 <= seconds < 60:
            break
        print("Invalid input. Check values and try again.")

if azimuth <= 90:
    # 0 is South, 90 is West. Moving from S toward W.
    bearing_angle = azimuth
    direction = f"S {bearing_angle:.2f}° W"

elif azimuth <= 180:
    # 180 is North, 90 is West. Moving from N toward W.
    bearing_angle = 180 - azimuth
    direction = f"N {bearing_angle:.2f}° W"

elif azimuth <= 270:
    # 180 is North, 270 is East. Moving from N toward E.
    bearing_angle = azimuth - 180
    direction = f"N {bearing_angle:.2f}° E"

else:
    # 360 is South, 270 is East. Moving from S toward E.
    bearing_angle = 360 - azimuth
    direction = f"S {bearing_angle:.2f}° E"

print(f"\nAzimuth: {azimuth:.2f}°")
print(f"Bearing: {direction}")
```

```python
benchmark_name = input("Enter the benchmark name: ")
benchmark_elevation = float(input("Enter the benchmark elevation: "))
level_setups = []
bs_point_name = benchmark_name


while True:
    print(f"\nBacksight point: {bs_point_name}")
    bs_reading = float(input("Enter BS reading: "))

    fs_point_name = input("Enter FS point name: ")
    fs_reading = float(input("Enter FS reading: "))
    
    level_setups.append({
        'bs_point': bs_point_name,
        'bs_reading': bs_reading,
        'fs_point': fs_point_name,
        'fs_reading': fs_reading
    })
    
    continue_choice = ""
    while continue_choice not in ['Y', 'N']:
        continue_choice = input("Continue? (Y/N): ").upper()
        
    if continue_choice == 'N':
        break
    
    bs_point_name = fs_point_name

print(f"\n{benchmark_name} {benchmark_elevation:.3f}")

current_elev = benchmark_elevation
for setup in level_setups:
    hi = current_elev + setup['bs_reading']
    new_elev = hi - setup['fs_reading']
    print(f"{setup['fs_point']} {new_elev:.3f}")
    current_elev = new_elev
```
